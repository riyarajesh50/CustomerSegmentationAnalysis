```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:/CustomerSegmentationAnalysis/data/ecommerce_customers.csv")

df.head()
```

```python
df.describe()

sns.histplot(df['Total Spend'])
plt.savefig("D:/CustomerSegmentationAnalysis/images/total_spend.png")
plt.show()

sns.boxplot(x=df['Items Purchased'])
plt.savefig("D:/CustomerSegmentationAnalysis/images/items_purchased.png")
plt.show()

sns.scatterplot(x='Total Spend', y='Items Purchased', data=df)
plt.savefig("D:/CustomerSegmentationAnalysis/images/scatter_chart.png")
plt.show()
```
